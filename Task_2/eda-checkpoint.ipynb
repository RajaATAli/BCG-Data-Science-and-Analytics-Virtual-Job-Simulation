
## Data visualization

Below are some functions that we can use to get started with visualizations. 

```python
def plot_stacked_bars(dataframe, title_, size_=(18, 10), rot_=0, legend_="upper right"):
    """
    Plot stacked bars with annotations
    """
    ax = dataframe.plot(
        kind="bar",
        stacked=True,
        figsize=size_,
        rot=rot_,
        title=title_
    )

    # Annotate bars
    annotate_stacked_bars(ax, textsize=14)
    # Rename legend
    plt.legend(["Retention", "Churn"], loc=legend_)
    # Labels
    plt.ylabel("Company base (%)")
    plt.show()

def annotate_stacked_bars(ax, pad=0.99, colour="white", textsize=13):
    """
    Add value annotations to the bars
    """

    # Iterate over the plotted rectanges/bars
    for p in ax.patches:
        
        # Calculate annotation
        value = str(round(p.get_height(),1))
        # If value is 0 do not annotate
        if value == '0.0':
            continue
        ax.annotate(
            value,
            ((p.get_x()+ p.get_width()/2)*pad-0.05, (p.get_y()+p.get_height()/2)*pad),
            color=colour,
            size=textsize
        )

def plot_distribution(dataframe, column, ax, bins_=50):
    """
    Plot variable distirbution in a stacked histogram of churned or retained company
    """
    # Create a temporal dataframe with the data to be plot
    temp = pd.DataFrame({"Retention": dataframe[dataframe["churn"]==0][column],
    "Churn":dataframe[dataframe["churn"]==1][column]})
    # Plot the histogram
    temp[["Retention","Churn"]].plot(kind='hist', bins=bins_, ax=ax, stacked=True)
    # X-axis label
    ax.set_xlabel(column)
    # Change the x-axis to plain style
    ax.ticklabel_format(style='plain', axis='x')
```

Thhe first function `plot_stacked_bars` is used to plot a stacked bar chart. An example of how you could use this is shown below:

```python
churn = client_df[['id', 'churn']]
churn.columns = ['Companies', 'churn']
churn_total = churn.groupby(churn['churn']).count()
churn_percentage = churn_total / churn_total.sum() * 100
plot_stacked_bars(churn_percentage.transpose(), "Churning status", (5, 5), legend_="lower right")
```

The second function `annotate_bars` is used by the first function, but the third function `plot_distribution` helps you to plot the distribution of a numeric column. An example of how it can be used is given below:

```python
consumption = client_df[['id', 'cons_12m', 'cons_gas_12m', 'cons_last_month', 'imp_cons', 'has_gas', 'churn']]

fig, axs = plt.subplots(nrows=1, figsize=(18, 5))

plot_distribution(consumption, 'cons_12m', axs)
```

Now, we will use the third function, `plot_distribution,` to help reveal if customers who churned were getting less of a discount then those who stayed

```python
fig, ax = plt.subplots()
plot_distribution(client_df, 'forecast_discount_energy', ax)
plt.show()
```

## Hypothesis Testing

To test the hypothesis that price sensitivity is correlated with churn, we first need to define price sensitivity. We might define a customer as being price sensitive if their consumption decreases when prices increase.

To measure this, we might create a new variable in the price dataframe that represents the change in price from the previous month

```python
price_df = price_df.sort_values(['id', 'price_date'])
price_df['price_change'] = price_df.groupby('id')['price_off_peak_var'].diff()
```

We might then merge this data with the client data

```python
df = pd.merge(client_df, price_df, how='left', on='id')
```

Next we can calculate the correlation between churn and price change

```python
correlation = df['churn'].corr(df['price_change'])
print('Correlation between churn and price change:', correlation)
```

```python
merged_df = pd.merge(client_df, price_df, on='id')

merged_df['price_sensitivity_off_peak'] = merged_df['cons_12m'] / merged_df['price_off_peak_var']
merged_df['price_sensitivity_peak'] = merged_df['cons_12m'] / merged_df['price_peak_var']
merged_df['price_sensitivity_mid_peak'] = merged_df['cons_12m'] / merged_df['price_mid_peak_var']

merged_df.replace([np.inf, -np.inf], np.nan, inplace=True)

merged_df.fillna(0, inplace=True)

merged_df['average_price_sensitivity'] = (merged_df['price_sensitivity_off_peak'] + merged_df['price_sensitivity_peak'] + merged_df['price_sensitivity_mid_peak']) / 3

correlation = merged_df[['churn', 'average_price_sensitivity']].corr()

print(correlation)

plt.figure(figsize=(10, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation heatmap between Churn and Average Price Sensitivity')
plt.show()
```
